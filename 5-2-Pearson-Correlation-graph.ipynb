```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, stats
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
#%%capture
df = pd.read_csv('data.csv')
#print(list(df.columns))
len(list(df.columns))
print(df['Age'])

# 1 Mild Cognitive Impairment
# 2 Dementia
# 3 Subjective Memory Complaint/Normal Cognition
# 4 Unknown - defer to record
```

```python
def draw_3d(df, x_column, y_column, z_column, color, label):

    # Create data points
    x = df[x_column]
    y = df[y_column]
    z = df[z_column]

    # Plot the surface
    ax.scatter(x,y,z, c=color, label = label)

    # Customize the plot
    ax.set_xlabel(x_column)
    ax.set_ylabel(y_column)
    ax.set_zlabel(z_column)
    ax.set_title('3D Scatter Plot')
    ax.legend()
```

```python
def draw_scatterplot(df, x_column, y_column, color, label):
   
    '''
    Draw scatterplot using data from a DataFrame.
    
    Parameters:
        df (DataFrame): input DataFrame containing the data.
        x_column (str): column  name for the x-axis data.
        y_column (str): column name for the y-axis data.
        color (str): color for the scatterplot points.
        label (str): label for the scatterplot points.
    
    Returns:
        None
    '''
    
    plt.scatter(df[x_column], df[y_column], c=color, label=label)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xlim(-50,100)
    plt.ylim(0,8000)
    plt.legend()
```

```python
# BNT_totalwstim_raw
CN_df = df[df['syndrome_v2_v2'] == 3]
CN_df = CN_df[['BNT_totalwstim_raw','Left-Caudate','Age']]
CN_df = CN_df.replace([np.nan, np.inf,-np.inf],np.nan).dropna()
CN_corr_r, p_value = pearsonr(CN_df['BNT_totalwstim_raw'], CN_df['Left-Caudate'])
print(f" CN: BNT vs. Left-Caudate r: {CN_corr_r}, p : {p_value}")

Dementia_df = df[df['syndrome_v2_v2'] == 2]
Dementia_df = Dementia_df[['BNT_totalwstim_raw','Left-Caudate','Age']]
Dementia_df = Dementia_df.replace([np.nan, np.inf,-np.inf],np.nan).dropna()
Dementia_corr, _ = pearsonr(Dementia_df['BNT_totalwstim_raw'], Dementia_df['Left-Caudate'])
print(f" Dementia: BNT vs. Left-Caudate r: {Dementia_corr}, p : {_}")


MCI_df = df[df['syndrome_v2_v2'] == 1]
MCI_df = MCI_df[['BNT_totalwstim_raw','Left-Caudate','Age']]
MCI_df = MCI_df.replace([np.nan, np.inf,-np.inf],np.nan).dropna()
MCI_corr, _ = pearsonr(MCI_df['BNT_totalwstim_raw'], MCI_df['Left-Caudate'])
print(f" MCI: BNT vs. Left-Caudate r: {MCI_corr}, p : {_}")

# BNT_totalwstim_raw vs. Left-Caudate
draw_scatterplot(CN_df, 'BNT_totalwstim_raw', 'Left-Caudate', 'green', 'CN')
draw_scatterplot(MCI_df, 'BNT_totalwstim_raw', 'Left-Caudate', 'blue', 'MCI')
draw_scatterplot(Dementia_df, 'BNT_totalwstim_raw', 'Left-Caudate', 'red', 'Dementia')

plt.show()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
draw_3d(CN_df, 'BNT_totalwstim_raw', 'Left-Caudate', 'Age', 'green', 'CN')
draw_3d(MCI_df, 'BNT_totalwstim_raw', 'Left-Caudate', 'Age', 'blue', 'MCI')
draw_3d(Dementia_df, 'BNT_totalwstim_raw', 'Left-Caudate', 'Age', 'red', 'Dementia')
# show the plot
plt.show()
```
