```python
import pandas as pd
from scipy.stats import levene
```

```python
# Levene_test
df = pd.read_csv('4-1-add-dx-lh_aparc.csv')
pd.set_option('display.max_columns', None)
#print(list(df.columns))

# 1 Mild Cognitive Impairment
# 2 Dementia
# 3 Subjective Memory Complaint/Normal Cognition
# 4 Unknown - defer to record

not_equal_var = {}
equal_var = {}

def levenetest(group_number1, group_number2, group_number3, interest):
    df1 = df[df['syndrome_v2_v2'] == group_number1]
    df1_interest = df1[interest]
    
    df2 = df[df['syndrome_v2_v2'] == group_number2]
    df2_interest = df2[interest]
    
    df3 = df[df['syndrome_v2_v2'] == group_number3]
    df3_interest = df3[interest]
      
    stat, p = levene(df1_interest, df2_interest, df3_interest)
    
    if p < 0.05:
        print("The variances are significantly different")
        not_equal_var[interest] = p
    else:
        print("The variances NOT different")
        equal_var[interest] = p
    print(f"{interest}: Levene's test statistic: {stat:.3f}, p-value{p:.3f}")
 
    
    
    
for ii in ['FAS_total_raw', 'FAS_total_T', 'Animals_raw', 'Animals_T',
       'BNT_totalwstim_raw', 'BNT_totalwstim_T', 'bankssts',
       'caudalanteriorcingulate', 'caudalmiddlefrontal', 'cuneus',
       'entorhinal', 'fusiform', 'inferiorparietal', 'inferiortemporal',
       'isthmuscingulate', 'lateraloccipital', 'lateralorbitofrontal',
       'lingual', 'medialorbitofrontal', 'middletemporal', 'parahippocampal',
       'paracentral', 'parsopercularis', 'parsorbitalis', 'parstriangularis',
       'pericalcarine', 'postcentral', 'posteriorcingulate', 'precentral',
       'precuneus', 'rostralanteriorcingulate', 'rostralmiddlefrontal',
       'superiorfrontal', 'superiorparietal', 'superiortemporal',
       'supramarginal', 'frontalpole', 'temporalpole', 'transversetemporal',
       'insula']:
    levenetest(1,2,3, ii)
```

```python
print('-'*50, 'NOT_equal_variance', '-'*45)
for key, value in not_equal_var.items():
    print(key, ' : ', value)
print('-'*50, 'equal_variance', '-'*50)
for key, value in equal_var.items():
    print(key, ' : ', value)
```
