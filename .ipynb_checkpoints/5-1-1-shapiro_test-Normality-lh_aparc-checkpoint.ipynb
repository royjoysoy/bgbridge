```python
import pandas as pd
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Shapiro_test
df = pd.read_csv('4-1-add-dx-lh_aparc.csv')
pd.set_option('display.max_columns', None)
#print(list(df.columns))

# 1 Mild Cognitive Impairment
# 2 Dementia
# 3 Subjective Memory Complaint/Normal Cognition
# 4 Unknown - defer to record

not_normal_MCI = []
not_normal_Dementia = []
not_normal_Normal = []

def shapirotest(group_number, interest):
    group_df = df[df['syndrome_v2_v2'] == group_number]
    df_interest = group_df[interest]
    w, p = shapiro(df_interest)      
    if group_number == 1:
        print("MCI")
        print(f'MCI:{interest}_w= {w:.3f}, {interest}_p {p:.3f}')
    elif group_number == 2:
        print("Dementia")
        print(f'Dementia:{interest}_w= {w:.3f}, {interest}_p {p:.3f}')
    elif group_number == 3:
        print("normal")
        print(f'Normal:{interest}_w= {w:.3f}, {interest}_p {p:.3f}')
    if p < 0.05:
        if group_number == 1:
            print("-"*15, "NOT NORMAL | MCI", interest, "-"*15)
            not_normal_MCI.append(interest)
        elif group_number == 2:
            print("-"*15, "NOT NORMAL | Dementia", interest, "-"*15)
            not_normal_Dementia.append(interest)
        elif group_number == 3:
            print("-"*15, "NOT NORMAL | NC", interest, "-"*15)
            not_normal_Normal.append(interest)

    else:
        print("normal")
        
        
    # create a density plot
    if group_number == 1:
        plt.hist(df_interest, density = True, alpha = 0.5)
        plt.title(f'{interest}')
        plt.xlabel(f'MCI,{interest}')
        plt.ylabel('Density')
        plt.show()
    
    if group_number == 2:
        plt.hist(df_interest, density = True, alpha = 0.5)
        plt.title(f'{interest}')
        plt.xlabel(f'Dementia,{interest}')
        plt.ylabel('Density')
        plt.show()
        
    if group_number == 3:
        plt.hist(df_interest, density = True, alpha = 0.5)
        plt.title(f'{interest}')
        plt.xlabel(f'CN,{interest}')
        plt.ylabel('Density')
        plt.show()
   
   
    
for i in [1, 2, 3]:
    for ii in ['FAS_total_raw', 'FAS_total_T', 'Animals_raw', 'Animals_T',
       'BNT_totalwstim_raw', 'BNT_totalwstim_T', 'bankssts',
       'caudalanteriorcingulate', 'caudalmiddlefrontal', 'cuneus',
       'entorhinal', 'fusiform', 'inferiorparietal', 'inferiortemporal',
       'isthmuscingulate', 'lateraloccipital', 'lateralorbitofrontal',
       'lingual', 'medialorbitofrontal', 'middletemporal', 'parahippocampal',
       'paracentral', 'parsopercularis', 'parsorbitalis', 'parstriangularis',
       'pericalcarine', 'postcentral', 'posteriorcingulate', 'precentral',
       'precuneus', 'rostralanteriorcingulate', 'rostralmiddlefrontal',
       'superiorfrontal', 'superiorparietal', 'superiortemporal',
       'supramarginal', 'frontalpole', 'temporalpole', 'transversetemporal',
       'insula']:
        shapirotest(i, ii)
```

```python
print('not_normal_MCI', not_normal_MCI,"\n")
print('not_normal_Dementia',not_normal_Dementia,"\n")
print('not_normal_Normal', not_normal_Normal)
```
