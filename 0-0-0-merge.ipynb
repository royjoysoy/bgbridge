```python
# add header to csv file
def add_header_to_csv(headers_dict):
    """
    Add header to multiple CSV files.
    
    Parameters:
        headers_dict (dict): Dictionary where keys are CSV file paths and values are lists of headers.
    """
    for csv_file_path, header in headers_dict.items():
        # Read existing data from CSV file
        existing_data = []
        with open(csv_file_path, mode='r', newline='') as file:
            reader = csv.reader(file)
            for row in reader:
                existing_data.append(row)
        
        # Prepend the headers to the data
        data_with_header = [header] + existing_data
        
        # Write the updated data, including the header, back to the CSV file
        with open(csv_file_path, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(data_with_header)
        
        print(f"Header added successfully to the CSV file: {csv_file_path}")

# Example usage:
# headers_dict = {
#     'data1.csv': ['Header1', 'Header2', 'Header3'],
#     'data2.csv': ['ID', 'Name', 'Age'],
#     'data3.csv': ['Timestamp', 'Value'],
# }

add_header_to_csv(headers_dict)
```

```python
#2. Convert txt to csv 
def txt_to_csv(input_files, output_folder):
    """
    Convert multiple text files to CSV files.
    
    Args:
    - input_files (list): List of input text file paths.
    - output_folder (str): Path to the output folder where CSV files will be saved.
    """
    for input_file in input_files:
        # Construct the output file path by replacing the extension
        output_file_path = os.path.join(output_folder, os.path.splitext(os.path.basename(input_file))[0] + ".csv")
        
        # Open the input text file for reading
        with open(input_file, 'r') as file:
            # Read lines from the input file
            lines = file.readlines()

        # Open the output CSV file for writing
        with open(output_file_path, 'w', newline='') as output_file:
            # Create a CSV writer object
            csv_writer = csv.writer(output_file)

            # Write each line from the input text file to the CSV file
            for line in lines:
                # Split the line into fields (assuming fields are separated by whitespace)
                fields = line.strip().split()  # Change split() argument as needed

                # Write the fields to the CSV file
                csv_writer.writerow(fields)

        print("Conversion complete. CSV file saved as:", output_file_path)
```
